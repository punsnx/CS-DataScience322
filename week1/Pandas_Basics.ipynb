```python
player_df[ player_df.team.str.endswith('land') ].head() # string method can be use if return boolean e.g. str.endswith
```

```python
df2 = player_df.sample(5,random_state=7) # random sample of 5 rows
df2
```

```python
df3 = df2.sample(4, random_state=8)
df3
```

```python
df2.passes + df3.passes 
# adding two dataframes base on index if index is student id 
# then if two dataframs not the same but still same index 
# it will add the value of the same index
# ! if index not the same it will return NaN or some dataframe don't have that index
# why value is become float because NaN is float
```

```python
df2.passes.add(df3.passes, fill_value=0) # add two dataframes and fill the missing value with 0
```
