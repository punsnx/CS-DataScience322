# 22.แสดงประเทศที่ได้เหรียญรวมทุกประเภทของโอลิมปิกฤดูหนาวมากกว่าเหรียญรวมโอลิมปิกฤดูร้อน ให้คอลัมน์ของตารางนี้แสดงชื่อประเทศ เหรียญรวมฤดูร้อน เหรียญรวมฤดูหนาว

```python
sum_season_medals = medals[['Country']].copy()
sum_season_medals['SummerMedals'] = medals.loc[: , medals.columns.str.contains('Summer(Gold|Silver|Bronze)')].sum(axis='columns')
sum_season_medals['WinterMedals'] = medals.loc[: , medals.columns.str.contains('Winter(Gold|Silver|Bronze)')].sum(axis='columns')
```

```python
sum_season_medals[ sum_season_medals.WinterMedals > sum_season_medals.SummerMedals ]
```

# 26.ให้สร้างคอลัมน์ใหม่ในตาราง medals ให้ชื่อว่า score โดยคำนวณคะแนนรวมจากเหรียญที่แต่ละประเทศได้(ทั้งฤดูร้อนและหนาว) โดยให้เหรียญทองได้ 3 คะแนน เหรียญเงินได้ 2 คะแนน เหรียญทองแดงได้ 1คะแนน แล้วแสดงข้อมูล 10 ประเทศแรกที่ได้คะแนนเยอะที่สุด

```python
medals['Score'] = (
(medals.loc[:,medals.columns.str.contains('Gold')].sum(axis='columns') * 3) + 
(medals.loc[:,medals.columns.str.contains('Silver')].sum(axis='columns') * 2) +
(medals.loc[:,medals.columns.str.contains('Bronze')].sum(axis='columns') * 1)
)
```

```python
medals.sort_values('Score',ascending=False).head(10)
```

# 27.การเปรียบเทียบคะแนนจากเหรียญทั้งหมดอาจจะไม่แฟร์ เพราะแต่ละประเทศเข้าร่วมการแข่งขันไม่เท่ากันดังนั้นให้สร้างคอลัมน์ใหม่ในตาราง medals ชื่อว่า perGame โดยคำนวณคะแนนต่อการเข้าร่วมแข่งขัน(คะแนนคำนวณเหมือนเดิม แต่หารด้วยจำนวนเกมส์ที่เข้าแข่งขัน แยกกันระหว่างฤดูร้อนกับฤดูหนาว ค่าperGame จะเป็นผลบวกของคะแนนของฤดูร้อนกับฤดูหนาว) แสดงข้อมูล 10 ประเทศที่ได้คะแนนน้อยที่สุด 
1. ระวัง ตรวจสอบผลลัพธ์ให้ดี มี surprise นะข้อนี้

```python
medals['ScoreSummer'] = (
(medals.loc[:,['SummerGold']].sum(axis='columns') * 3) + 
(medals.loc[:,['SummerSilver']].sum(axis='columns') * 2) +
(medals.loc[:,['SummerBronze']].sum(axis='columns') * 1)
)
```

```python
medals['ScoreWinter'] = (
(medals.loc[:,['WinterGold']].sum(axis='columns') * 3) + 
(medals.loc[:,['WinterSilver']].sum(axis='columns') * 2) +
(medals.loc[:,['WinterBronze']].sum(axis='columns') * 1)
)
```

```python
medals['PerGameSummer'] = medals.ScoreSummer.div(medals.SummerGame)
medals['PerGameWinter'] = medals.ScoreWinter.div(medals.WinterGame)
```

```python
medals['PerGameSummer'].fillna(0,inplace=True)
medals['PerGameWinter'].fillna(0,inplace=True)  
```

```python
medals['PerGame'] = medals.PerGameSummer + medals.PerGameWinter
```

```python
medals.sort_values('PerGame',ascending=True).head(10).loc[:,medals.columns.str.contains('(^Country$|PerGame)')]
```
