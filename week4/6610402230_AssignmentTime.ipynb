# 1.เปิดไฟล์ italy_earthquakes_from_2016-08-24_to_2016-11-30.csv

```python
import pandas as pd
```

```python
df = pd.read_csv('italy_earthquakes_from_2016-08-24_to_2016-11-30.csv')
df
```

# 2.ปรับคอลัมน์ Time ให้เป็น Timestamp และตั้งให้เป็น index

```python
df.Time = pd.to_datetime(df.Time)
df.set_index('Time',inplace=True)
df
```

# 3. เวลาที่เกิดแผ่นดินไหวแรงที่สุดคือวันเวลาไหน

```python
max_magnitude_date = df[df['Magnitude'] == df['Magnitude'].max()].index[0]
max_magnitude_date.strftime('%A %b %dth, year %Y Time %H : %M : %S')
```

 # 4. ในวันที่เกิดแผ่นดินไหวแรงที่สุด มีแผ่นดินไหวกี่ครั้ง

```python
max_magnitude_date_str = max_magnitude_date.strftime('%Y-%m-%d')
df.loc[max_magnitude_date_str].shape
```

# 13. สมมติว่าเรานิยามแผ่นดินไหวที่รุนแรงนั้นต้องมี Magnitude มากกว่าการไหวครั้งก่อนและครั้งหลังอย่างน้อย 2.0 จงคัดเลือกเหตุการณ์แผ่นดินไหวที่รุนแรงออกมา

```python
df.Magnitude.diff(1)
```

```python
df.Magnitude.diff(-1)
```

```python
high_magnitude_df = df[((df.Magnitude.diff(1) >= 2) & (df.Magnitude.diff(-1) >= 2))]
high_magnitude_df
```

# 14. จากตารางข้อ 13. ให้เพิ่มคอลัมน์ใหม่ชื่อ hour_count แสดงจำนวนแผ่นดินไหวในชั่วโมงที่มีเหตุการณ์แผ่นดินไหวรุนแรงด้วย

```python
df.resample('1H').size().to_period()
```

```python
time_high_magnitude = high_magnitude_df.index[0].strftime('%Y-%m-%d %H:%M:%S')
time_high_magnitude
```

```python
hourly_counts = df.resample('1H').size()
high_magnitude_df['hour_count'] = high_magnitude_df.index.to_series().apply(lambda x: hourly_counts[x.floor('H')])

high_magnitude_df
```

# 16. คำนวณค่าเฉลี่ยของความรุนแรงแผ่นดินไหวที่มากที่สุดในช่วง 3 วัน โดยหาความแรงของแผ่นดินไหวที่มากที่สุดในแต่ละวันก่อน แล้วใช้ sliding window หาค่าเฉลี่ยของทุก ๆ 3 วัน

```python
df.Magnitude.resample('1D').max()
```

```python
for window in df.Magnitude.resample('1D').max().rolling('3D'):
    print(window,"\nmean : ",window.mean(),'\n')
```
