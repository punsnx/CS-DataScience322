# 6. นำคอลัมน์ที่ชื่อ continent ของ drinks ไปต่อกับตาราง medals ตามชื่อประเทศของ medals

```python
df_drinks.rename(columns={'country':'Country'},inplace=True)
df_drinks
```

```python
df_medals_with_continent = df_medals.join(df_drinks['continent'],how='left')
print(df_medals.shape,df_medals_with_continent.shape)
df_medals_with_continent
```

# 7. ในตาราง medals จงแสดงชื่อประเทศที่ยังไม่มีข้อมูลในคอลัมน์ continent

```python
df_medals_with_continent[df_medals_with_continent['continent'].isnull()].index.unique()
```

# 8. นิสิตจะเห็นแนวทางการแก้ไขข้อมูลหรือรวมประเทศบางประเทศเข้าด้วยกัน แต่ในการบ้านนี้ยังไม่ต้องทำทั้งหมด แค่เติมข้อมูลทวีปที่ยังขาดหายไปในบางประเทศดังนี้ Australasia: OC, Canada/United States/Mexico: NA, Cuba/Jamaica: SA, Czech/Yugoslavia/Germany(ทั้งหมด)/Great Britain/Russia(ทั้งหมด)/Soviet: EU ประเทศที่เหลือกำหนดให้เป็น 'OT' ไป
1. ตารางนี้จะเป็นตารางแบบ wide ที่สมบูรณ์ ควรเก็บไว้เพื่อใช้ในการคำนวณ

```python
df_medals_with_continent[df_medals_with_continent['continent'].isnull()].shape
```

```python
fill_country = {'Australasia':'OC','Canada':'NA','United States':'NA','Mexico':'NA','Czech':'EU','Yugoslavia':'EU','Germany':'EU','Great Britain':'EU','Russia':'EU','Soviet':'EU'}
for key in fill_country.keys():
    df_medals_with_continent.loc[df_medals_with_continent.index.str.contains(key),'continent'] = fill_country[key]
```

```python
df_medals_with_continent[df_medals_with_continent['continent'].isnull()].shape
```

```python
df_medals_with_continent.loc[df_medals_with_continent['continent'].isnull(),'continent'] = 'OT'
```

```python
df_medals_with_continent[df_medals_with_continent['continent'].isnull()].shape
```

```python
print(df_medals_with_continent.info())
df_medals = df_medals_with_continent
```

# 9. ตั้งให้ ชื่อทวีปเป็น index ชั้นนอกและชื่อประเทศเป็น index ชั้นใน

```python
df_medals.reset_index(inplace=True)
df_medals.set_index(['continent','Country','Code'],inplace=True)
df_medals
```

# 10. เรียงลำดับชื่อทวีปจากมากไปน้อยและในทวีปเรียงลำดับชื่อประเทศจากน้อยไปมาก

```python
df_medals.sort_index(ascending=[False,True,True],inplace=True)
df_medals
```

# 11. จัดคอลัมน์ให้เป็นสองระดับ โดยระดับแรกคือฤดู และระดับที่สองคือจำนวนเกมส์และเหรียญที่ได้ (ดูตัวอย่างจากเอกสารการสอน) ดังนั้นตอนนี้ข้อมูลจะมีสองระดับทั้ง index (ทวีปและประเทศ) และ column(ฤดูและจำนวนเกมและเหรียญ)

```python
df_medals_multi_index_cols = pd.MultiIndex.from_tuples([
    ('Summer','SummerGame'),
    ('Summer','SummerGold'),
    ('Summer','SummerSilver'),
    ('Summer','SummerBronze'),
    ('Winter','WinterGame'),
    ('Winter','WinterGold'),
    ('Winter','WinterSilver'),
    ('Winter','WinterBronze'),
])
old_df_medals_cols = df_medals.columns
```

```python
# df_medals.columns = old_df_medals_cols
```

```python
df_medals.columns = df_medals_multi_index_cols
df_medals
```

# 12. ย้ายระดับของฤดูมาเป็น index ชั้นในสุด

```python
df_medals.stack(level=0)
```

# 13. ปรับตารางให้เป็นแบบ long โดยหนึ่งแถวจะมีคอลัมน์คือทวีป ประเทศ ฤดูแข่งขัน จำนวนเกม ประเภทเหรียญ จำนวนเหรียญ

```python
df_medals.columns = old_df_medals_cols
df_melt = df_medals.reset_index()
df_melt
```

```python
df_melt.columns
```

```python
df_melt = df_melt.melt(id_vars=['continent', 'Country', 'Code', 'SummerGold',
       'SummerSilver', 'SummerBronze', 'WinterGold',
       'WinterSilver', 'WinterBronze'],value_vars=['SummerGame','WinterGame'],var_name='Season',value_name='Game')
```

```python
df_melt
```

```python
df_melt = df_melt.melt(id_vars=['continent', 'Country', 'Code', 'Season','Game'],value_vars=['SummerGold',
       'SummerSilver', 'SummerBronze', 'WinterGold',
       'WinterSilver', 'WinterBronze'],var_name='Medal',value_name='Count')
```

```python
df_melt
```

```python
df_melt['Season'] = df_melt['Season'].str[:6]
```

```python
df_melt['Medal'] = df_melt['Medal'].str[6:]
```

```python
df_melt = df_melt[['continent', 'Country', 'Code', 'Season', 'Game', 'Medal', 'Count']]
df_melt
```

```python
df_melt[df_melt['Country'] == 'Austria']
```

# 15. นับว่าแต่ละทวีปมีกี่ประเทศ (ลองใช้คำสั่ง size() ต่อจากการ groupby() )

# 16. แสดงตารางรวมเหรียญฤดูร้อนแต่ละประเภทของทุกประเทศในทวีปเดียวกัน แสดงผลแยกตามทวีป

```python
df_medals.loc[:,df_medals.columns.str.contains('continent|Summer')].groupby('continent').sum()
```

# 17. แสดงตารางสรุปรวมเหรียญแต่ละประเภทของทุกประเทศแยกตามฤดูแข่งขัน

```python
df_melt.filter(['Season','Medal','Count']).pivot_table(index='Medal',columns='Season',values='Count',aggfunc='sum')
```

# 18. หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูร้อนต่อประเทศที่เข้าร่วมแข่งขัน แยกตามทวีป

```python
df_medals.loc[:,df_medals.columns.str.contains('continent|SummerGold')].groupby('continent').mean()
```

# 19. หาจำนวนประเทศที่เคยเข้าร่วมการแข่งขันโอลิมปิค แยกตามทวีปและฤดูแข่งขัน แสดงในตารางเดียวกัน

```python
df_melt.filter(['continent','Season','Game']).pivot_table(index='continent',columns='Season',values='Game',aggfunc='count')
```
