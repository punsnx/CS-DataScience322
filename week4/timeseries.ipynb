```python
kc_house_df = pd.read_csv('kc_house_data.csv')
kc_house_df
```

## date in df to datetime

```python
kc_house_df['date'] = pd.to_datetime(kc_house_df.date)
kc_house_df.head()
```

## set index as timestamp

```python
kc_house_df.set_index('date',inplace=True)
```

```python
kc_house_df.sort_index()
```

```python
kc_house_df.head()
```

## Slicing

```python
kc_house_df.loc['2015-05':]
```

## Resample

```python
kc_house_df.resample('M').size()
```

```python
kc_house_df.resample('2W-SAT').price.mean()
```

```python
kc_house_df.resample('ME').size()
```

## Period

```python
kc_house_df.resample('ME').size().to_period()
```

```python
kc_house_df.resample('Q').size().to_period()
```

## different func

```python
monethly_house_sale = kc_house_df.resample('ME').size()
monethly_house_sale
```

```python
monethly_house_sale.to_period()
```

## diff function

```python
monethly_house_sale.diff()
```

## shifting

```python
monethly_house_sale.to_period().shift()
```

```python
pd.concat([monethly_house_sale.to_period(),monethly_house_sale.to_period().shift(3)], axis='columns')
```

### shifting by time

```python
monethly_house_sale.shift(freq='3D') #3days
```

## windowing 

```python
weekly_house_sale =  kc_house_df.resample('W').size()
```

```python
weekly_house_sale.head()
```

if we want to mean for every 3 days we need to use windowing the data by 3

```python
sliding_window = weekly_house_sale.rolling(3)
sliding_window
```

the row 1,2 is NaN because of not complete windows then in row 3
we can windowing 3 row and mean them

```python
sliding_window.mean()
```
