# Seaborn 

Seaborn is from matplotlib but easy to use
- Exploratory data analysis

```python
import pandas as pd
import numpy as np

df = pd.read_csv("tips.csv")
df
```

```python
df.tip.mean(),df.tip.std()
```

```python
import seaborn as sns
```

## distribution plot

```python
sns.displot(df.tip,kind='hist',height=3)
```

```python
sns.displot(df.tip,kind='hist',height=3,bins=10)
```

- figure-level (whole large image)
- axis-level (sub-level)

```python
sns.displot(df.tip,kind='kde',height=3,aspect=1.5)
```

```python
sns.displot(df.tip,kind='kde',height=3,aspect=2,rug=True)
```

## relation plot

```python
sns.relplot(df,x='total_bill',y='tip',height=3,aspect=1.5,hue='sex')
```

```python
sns.relplot(df,x='total_bill',y='tip',height=3,aspect=1.5,col='sex')
```

```python
sns.relplot(df,x='total_bill',y='tip',height=3,aspect=1.5,row='sex')
```

```python
sns.relplot(df,x='total_bill',y='tip',height=3,aspect=1.5,col='sex',row='time')
```

```python
sns.relplot(df,x='total_bill',y='tip',height=3,aspect=1.5,style='sex')
```

must use in explicitly different data size

```python
sns.relplot(df,x='total_bill',y='tip',height=3,aspect=1.5,size='size')
```

### customize for kde

```python
sns.displot(df,x='tip',kind='kde',height=3,aspect=1.5,hue='sex')
```

```python
sns.displot(df,x='tip',kind='kde',height=3,aspect=1.5,hue='sex',palette='prism')
```

## category plot

```python
df['day_cat'] = pd.Categorical(df.day,categories=['Thur','Fri','Sat','Sun'],ordered=True)
```

default of bar is mean

```python
sns.catplot(df,x='day_cat',y='tip',height=3,kind='bar')
```

```python
sns.catplot(df,x='day_cat',y='tip',height=3,kind='bar',estimator='sum')
```

```python
sns.catplot(df,x='day_cat',y='tip',height=3,kind='point')
```

## pair plot

```python
sns.pairplot(df,height=3)
```

## pair grid

```python
g = sns.PairGrid(df,hue='sex')
g.map_diag(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.barplot)
```

## set style

```python
sns.set_style('dark')
sns.set_context('paper')
g = sns.PairGrid(df,hue='sex')
g.map_diag(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.barplot)
```
