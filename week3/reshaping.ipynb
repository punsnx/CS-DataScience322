### loc multi_index d-1

## multi columns index

```python
x = pd.MultiIndex.from_tuples([('Info','position'),
                           ('Info','minutes'),
                           ('Attack','shots'),
                           ('Attack','passes'),
                           ('Defense','tackles'),
                           ('Defense','saves')])
x
```

```python
multi_index_player.columns = x
multi_index_player
```

### finding index level

```python
multi_index_player.xs('Howard',level=1)
```

```python
multi_index_player.xs('Algeria',level=0)
```

# Multi level sorting

## ascending 1-level

```python
multi_index_player.sort_index(ascending=False)
```

## ascending multi-level

```python
multi_index_player.sort_index(ascending=[True,True])
```

# Swap index level

```python
multi_index_player.swaplevel()
```

## swap multi-level index

```python
df_3_level = player_df.set_index(['team','position','surname'])
df_3_level
```

if we have multi-index level the swaplevel() function swap from the deepest level to previous 

```python
df_3_level.swaplevel()
```

## specify the level for swaplevel()

normally we swaplevel with sort index

```python
df_3_level.swaplevel(i=0,j=1).sort_index()
#this is the chaining swaplevel().sort_index()
```

# Pivot / melt

```python
eu_city_temp = pd.read_csv('eu_city_temp.csv',skiprows=1).loc[:,:"Dec"]
eu_city_temp.head()
```

## melt()
unpivot the columns in to one column variable with its value columns

this function useful for preprocessing data in machine learning

- Long mean many rows
- Wide mean many columns

```python
long_eu_temp = eu_city_temp.melt(id_vars=['Country','City'],var_name='Month',value_name='Temperature')
long_eu_temp
```

## pivot()

**note the data index will sort ascending

```python
wide_eu_temp = long_eu_temp.pivot(index=['Country','City'],columns='Month',values='Temperature')
wide_eu_temp.head()
```

# Group by

In pandas group by do the split-apply-combine
- similar to divide and conquer

SPLIT STEP
- e.g. if we do g = dataframe.groupby(columns)
- pandas make each value in columns into sub table with another columns

APPLY & COMBINE 
- g.columns.sum()
- apply all group with sum columns then combine to one columns

```python
player_df.passes.sum()
```

```python
player_df
```

```python
player_df.query('team=="Algeria"').passes.sum()
```

```python
g = player_df.groupby('position')
```

```python
g.groups
```

```python
g.passes.sum()
```

```python
player_df.groupby('team').passes.mean().head(5)
```

# pivot table

if there are duplicate value before 
- the pivot not allow to do because the value from Ctype will duplicate 
- the pivot table have argrument that is aggregation function
  
Let see the example 

**note if the value not exit still can cause nan value

```python
player_df.pivot_table(index='team',columns='position',values='passes',aggfunc='sum').head()
```
